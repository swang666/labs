### practice

```python
# 1. create a year column from date column
# 2. group by ticker, get the average return
# 3. group by ticker and year, get the average return










```

```python
df['date'] = pd.to_datetime(df['date'], format = '%Y%m%d')
```

```python
import datetime
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
df.groupby('TICKER')['RET'].mean()
```

```python
df.groupby(['TICKER', 'year'])['RET'].mean()
```

```python
df2 = df.groupby(['TICKER', 'year']).agg({
    'RET': ['mean', 'std', lambda x: np.mean(x)/np.std(x)]
}).rename(columns={'<lambda>':'SR'})
```

```python
df2.columns = df2.columns.droplevel(0)
df2 = df2.reset_index()
df2
```

```python
# merge 2 dataframes
df.merge(df2, on = ['TICKER','year'])
```

```python
df['lag_ret'] = df.groupby(['TICKER'])['RET'].shift(1)
```

```python
# a momentum strategy
df['11_month_moving_average'] = 0
for i in range(2,13):
    df['11_month_moving_average'] = df['11_month_moving_average'] + df.groupby(['TICKER'])['RET'].shift(i)
df['11_month_moving_average'] = df['11_month_moving_average']/11
```

```python
df['max_mom'] = df.groupby(['year','month'])['11_month_moving_average'].transform(max)
df['min_mom'] = df.groupby(['year','month'])['11_month_moving_average'].transform(min)
```

```python
long_leg = df[df['11_month_moving_average'] == df['max_mom']].sort_values(by=['year', 'month'])
short_leg = df[df['11_month_moving_average'] == df['min_mom']].sort_values(by=['year', 'month'])
```

```python
port = long_leg['RET'].values - short_leg['RET'].values
```

```python
annualized_mean = port.mean()*12
annualized_mean
```

```python
annualized_std = port.std()*np.sqrt(12)
annualized_std
```

```python
SR = annualized_mean/annualized_std
SR
```
